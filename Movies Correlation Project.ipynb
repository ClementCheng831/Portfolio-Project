```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv('/Users/clementcheng/Downloads/movies.csv')
```

```python
# First look at the data

df.head(10)
```

```python
# See if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Remove missing data from the data frame

df = df.dropna()
```

```python
# Double check if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Remove any duplicates

df = df.drop_duplicates()
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of the columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
# Preview data after changing the data type

df.head(10)
```

```python
# Create Correct Year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)


# See if the yearcorrect columns works
# Especialy look at row 8

df.head(10)
```

```python
# Sorting data by gross in descending order

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Allow us to observe all rows from our csv files

pd.set_option('display.max_rows', None)
```

```python
# My hypnosis

# H1: Budget has high correlation with gross

# H2: Company has high correlation with gross

# H3: Genre has high correlation with gross
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')

plt.show()

```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "black"}, line_kws={"color": "red"})
```

```python
# Looking at correlation

df.corr(method='pearson') # pearson, kendall, spearman
```

```python
# As the correlation coefficient value is positive and close to 1

# High correlation between budget and gross

# H1 was correct
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Numerized our data columns for further analysis on scatter plot

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(10)
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Quicker way to see the correlation

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the above table

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Extract high correlation movies features

high_cor = sorted_pairs[(sorted_pairs) > 0.5]

high_cor
```
