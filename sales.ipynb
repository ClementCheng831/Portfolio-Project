```python
# Modules

import pandas as pd
from datetime import datetime
```

```python
# Have a quick look of the data

df = pd.read_csv('/Users/clementcheng/Downloads/Supermarket/supermarket_sales - Sheet1.csv')

df.head(5)
```

```python
# Check if the data is clean
# 1. Check for null values

df.isna().sum()
```

```python
# Check at the data type

df.dtypes
```

```python
# We need to change the Date & Time column into datetime 

df['Date'] = df['Date'].astype('datetime64')

df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
```

```python
# Add Month and Year column for better filter

df['Month'] = df['Date'].dt.month

df['Year'] = df['Date'].dt.year
```
